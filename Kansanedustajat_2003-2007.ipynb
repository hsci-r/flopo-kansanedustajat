```python
import requests
from bs4 import BeautifulSoup

URL = "https://fi.wikipedia.org/wiki/Luettelo_vaalikauden_2003%E2%80%932007_kansanedustajista"
page = requests.get(URL)

soup = BeautifulSoup(page.content, "html.parser")
```

```python
tabs = soup.find_all("table",{"class":"wikitable sortable"})
ked = tabs[0]
rows = ked.find_all("tr")
```

```python
i = 0
names = []
parties = []
for row in rows:
    if i == 0:
        i += 1
        continue
    columns = row.find_all("td")
    #print(columns)
    name = columns[0].find_all("a")[0].get_text()
    print(name)
    name = name.split(" ")[1] + ", " + name.split(" ")[0]
    party = columns[1].find_all("a")[0].get_text()
    print(name, party)
    names.append(name)
    parties.append(party)
    i += 1
    
print(i)
```

```python
import pandas as pd
df = pd.DataFrame(list(zip(names, parties)),
               columns =['name', 'party'])

df.to_csv("kansanedustajat_2003-2007.csv")
```
