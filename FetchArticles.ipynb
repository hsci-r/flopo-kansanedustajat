```python
import pandas as pd
import os

file_list = ['data/kansanedustajat_1999-2003.csv', 'data/kansanedustajat_2003-2007.csv', 'data/kansanedustajat_2007-2011.csv', 'data/kansanedustajat_2011-2015.csv', 'data/kansanedustajat_2015-2019.csv']
```

```python
import requests

failed = []
unique_names = set()

for file in file_list:
    df = pd.read_csv(file)
    folder_files = os.listdir("document_ids")
    print(df.columns)
    
    session = requests.Session()
    session.auth = ("flopo", "dhh17")
    #print(file)
    
    for index, row in df.iterrows():
        firstname = row['firstname'].replace("-", "%23")
        lastname = row['lastname'].replace("-", "%23")
        name = firstname + "_" + lastname + ".csv"
        unique_names.add(name)
        #name = name.replace("-", "#")
        #print(name)
        yle_query = "https://flopo-yle-octavo.rahtiapp.fi/yle/search?query=%22L%3D" + firstname + "%20L%3D" + lastname + "%22&limit=-1&field=document_id&format=CSV"
        #print(yle_query)
        hs_query = "https://flopo-hs-octavo.rahtiapp.fi/hs/search?query=%22L%3D" + firstname + "%20L%3D" + lastname + "%22&limit=-1&field=document_id&format=CSV"
        #print(hs_query)
        stt_query = "https://flopo-stt-octavo.rahtiapp.fi/stt/search?query=%22L%3D" + firstname + "%20L%3D" + lastname + "%22&limit=-1&field=document_id&format=CSV"
        #print(stt_query)
        il_query = "https://flopo-il-octavo.rahtiapp.fi/il/search?query=%22L%3D" + firstname + "%20L%3D" + lastname + "%22&limit=-1&field=document_id&format=CSV"
        #print(il_query)
        if name not in os.listdir("document_ids/yle"):
            try:
                r = session.get(yle_query, allow_redirects=True)
                if r.status_code == 200:
                    open('document_ids/yle/' + name, 'wb').write(r.content)
                    if len(str(r.text).split("\n")) <= 2:
                        print(yle_query)
                        print(name)
                        print(r.text)
                else:
                    print("Error ",  r.status_code)
                    failed.append(yle_query)
            except:
                print("Error")
                failed.append(yle_query)
        if name not in os.listdir("document_ids/hs"):
            try:
                r = session.get(hs_query, allow_redirects=True)
                if r.status_code == 200:
                    open('document_ids/hs/' + name, 'wb').write(r.content)
                    if len(str(r.text).split("\n")) <= 2:
                        print(hs_query)
                        print(name)
                        print(r.text)
                else:
                    print("Error ",  r.status_code)
                    failed.append(hs_query)
            except:
                print("Error")
                failed.append(hs_query)
        if name not in os.listdir("document_ids/stt"):
            r = session.get(stt_query, allow_redirects=True)
            try:
                if r.status_code == 200:
                    open('document_ids/stt/' + name, 'wb').write(r.content)
                    #print(r.text)
                    if len(str(r.text).split("\n")) <= 2:
                        print(stt_query)
                        print(name)
                        print(r.text)
                else:
                    print("Error ",  r.status_code, " ", stt_query)
                    failed.append(stt_query)
            except:
                print("Error")
                failed.append(stt_query)
        if name not in os.listdir("document_ids/il"):
            try:
                r = session.get(il_query, allow_redirects=True)
                if r.status_code == 200:
                    open('document_ids/il/' + name, 'wb').write(r.content)
                    if len(str(r.text).split("\n")) <= 2:
                        print(il_query)
                        print(name)
                        print(r.text)
                else:
                    print("Error ", r.status_code)
                    failed.append(il_query)
            except:
                print("Error")
                failed.append(il_query)
            
            
# # https://flopo-il-octavo.rahtiapp.fi/il/search
print(len(unique_names))
```
