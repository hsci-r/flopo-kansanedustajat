```python
import pandas as pd

import requests
from bs4 import BeautifulSoup

URL = "https://valtioneuvosto.fi/tietoa/historiaa/hallitukset-ja-ministerit/raportti/-/r/m1/74"
page = requests.get(URL)

soup = BeautifulSoup(page.content, "html.parser")

```

```python
tabs = soup.find_all("tbody")
ked = tabs[0]
rows = ked.find_all("tr")

names = []
nimikkeet = []

i = 0
while True:
    nimike = rows[i]
    nimike = nimike.find("strong").get_text()
    
    j = i + 1
    while True:
        if "strong" in str(rows[j]):
            break
        else:
            nimi = rows[j]
            nimi = nimi.find_all("td")[0].get_text()
            print(nimi, nimike)
            names.append(nimi)
            nimikkeet.append(nimike)
        #print(j+1)
        #print(len(rows)-1)
        if j +1 > len(rows)-1:
            j += 1
            break
        j += 1
    i = j
    if i > len(rows)-1:
        break
    
```

```python
print(nimikkeet)
print(names)
nimikkeet_dict = {}

for i in range(0, len(names)):
    if names[i] not in nimikkeet_dict:
        nimikkeet_dict[names[i]] = nimikkeet[i]
    else:
        nimikkeet_dict[names[i]] = nimikkeet_dict[names[i]] + ";; " + nimikkeet[i]
        
print(nimikkeet_dict)

df_nimikkeet = []

filename = "kansanedustajat_2015-2019.csv"
df = pd.read_csv(filename)
i=0

for index,row in df.iterrows():
    firstname = row['firstname']
    lastname = row['lastname']
    a = " "
    if 'position' in df.columns:
        a = row['position'] 
    for k in nimikkeet_dict.keys():
        if firstname in k and lastname in k:
            print(k)
            print(a)
            a += nimikkeet_dict[k] + ";;"
    df_nimikkeet.append(a)
    
        
df['position'] = df_nimikkeet
df = df[['name','party', 'firstname', 'lastname', 'position']]
df.to_csv(filename)
```
