```python
import os
import pandas as pd

names = []
parties = []
i = 0
for file in os.listdir("1999-2003"):
    if ".txt" in file:
        content = open("1999-2003/" + file, "r").read()
        party = file.split(".")[0]
        for row in content.split("\n"):
            name = row.split(",")[0]
            name = name.split(" ")[1] + ", " + name.split(" ")[0]
            print(name, party)
            names.append(name)
            parties.append(party)
            i += 1
            
print(i)
```

```python
import pandas as pd
df = pd.DataFrame(list(zip(names, parties)),
               columns =['name', 'party'])

df.to_csv("kansanedustajat_1999-2003.csv")
```
