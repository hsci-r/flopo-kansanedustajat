```python
import pandas as pd

file_list = ['kansanedustajat_1999-2003.csv', 'kansanedustajat_2003-2007.csv', 'kansanedustajat_2007-2011.csv', 'kansanedustajat_2011-2015.csv', 'kansanedustajat_2015-2019.csv']
years = ['1999-2003', '2003-2007', '2007-2011', '2011-2015', '2015-2019']
dfs = []

i = 0
for file in file_list:
    df = pd.read_csv(file)
    df['years'] = years[i]
    dfs.append(df)
    i += 1
```

```python
joined = pd.concat(dfs)

joined
```

```python
print(set(joined['party']))

joined["party"].replace({"vihr": "Vihreät", "kd": "Kristillisdemokraatit", "sdp": "SDP", 
                    "vas": "Vasemmistoliitto", "rkp": "RKP", "Sosiaalidemokraatit": "SDP", "ps": "Perussuomalaiset",
                        "perussuomalaiset":"Perussuomalaiset", "sd":"SDP", "r":"RKP", "kesk":"Keskusta",
                        "kokoomus":"Kokoomus", "vasemmistoliitto":"Vasemmistoliitto", "vr":"Vasenryhmä",
                        "keskusta":"Keskusta","vihreät":"Vihreät", "kok":"Kokoomus"}, inplace=True)

print(set(joined['party']))
```
